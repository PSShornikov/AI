## Импортируем библиотеки

```python
import cv2
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
matplotlib.rcParams['figure.figsize'] = (10,10)
```

## Импортируем изображение

```python
im = cv2.imread("/Users/dmitry/dergach.jpg")

plt.imshow(im)

print("Размер изображения:", str(im.shape[0]) + 'x' + str(im.shape[1]))
```

На изображении 50 колонок и 10 строка. По 2 строкb на каждый из классов.

Будем брать изображения по 32 пикселя.

```python
step = 32
line_num = 50
col_num = 10

step_by_class = 5

classes = ['A', 'B', 'D', 'E','G']
```

## Изменим размер изображения

```python
need_height = col_num*step
print("Необходимый размер в высоту:", need_height)

need_weight = line_num*step
print("Необходимый размер в ширину:", need_weight)

resized = cv2.resize(im, (need_weight, need_height), interpolation = cv2.INTER_AREA)

plt.imshow(resized)

print("Размер изображения:", str(resized.shape[0]) + 'x' + str(resized.shape[1]))

plt.imshow(resized[1:step-1, 1:step-1])
```

## Вырежем изображения

```python
plt.imshow(resized[1:step-1, 1:step-1])


for n in np.arange(len(classes)):
    for i in np.arange(n * step * step_by_class, (n + 1) * step * step_by_class, step):
        for j in np.arange(0, need_weight, step):
            cut = resized[i + 1 : i + step - 1, j + 1 : j + step - 1]
            cv2.imwrite("/Users/dmitry/dergach/" + classes[n] + "/img_" + str(i//step) + "x" + str(j//step) + ".jpg", cut)
```
